```python
sns.barplot(x = "Marital_Status",  y= "Loan_Default_Risk", data = loan)
plt.ylabel("Loan Default Risk")
plt.xlabel("Marital Status")
plt.title("Marital status vs Loan Default Risk")
plt.show()
```

The single people are mostly likely to default loan with the risk at 0.13 while the married one with the risk at 0.11.

```python
sns.barplot(x = "House_Ownership",  y= "Loan_Default_Risk", data = loan)
plt.ylabel("Loan Default Risk")
plt.xlabel("House Ownership")
plt.title("House Ownership vs Loan Default Risk")
plt.show()
```

Those in the rented houses are more likely to default the loan with the risk at 0.13 while those with owned and not rented or owned at the risk of 0.09

```python
sns.barplot(x = loan["Vehicle_Ownership(car)"], y= "Loan_Default_Risk", data = loan)
plt.ylabel("Loan Default Risk")
plt.xlabel("Vehicle Ownership")
plt.title("Vehicle Ownership vs Loan Default Risk")
plt.show()
```

Those without cars are likely to default loan with the risk slighlty higher than 0.14
